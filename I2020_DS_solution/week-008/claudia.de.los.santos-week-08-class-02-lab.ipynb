# Exercise 2

Create a function that will take a list of web pages and return a list of all the anchor tags on those pages.

```python
import requests

def scrape_web_pages(web_pages):
    anchor_tags_list = []
    for url in web_pages:
        request_response = requests.get(url)
        soup = BeautifulSoup(request_response.text)
        anchor_tags_list.extend(soup.find_all("a"))
    return anchor_tags_list

scrape_web_pages(['https://etsy.com', 'https://pinterest.com', 'https://cdc.com'])
        
        
```

# Exercise 3:

Create a function that accesses a web page and returns data from that page.

```python
def access_data(web_page):
    request_response = requests.get(web_page)
    soup = BeautifulSoup(request_response.text)
    data = soup.ul
    return data.contents[0].contents[0].contents


    
access_data("https://www.governor.ny.gov/news/governor-cuomo-updates-new-yorkers-states-progress-during-covid-19-pandemic-63")
```
